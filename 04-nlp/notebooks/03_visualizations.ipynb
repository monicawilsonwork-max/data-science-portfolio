In this lesson, we'll practice:

- Creating line charts, bar charts, and histograms
- Labeling axes and adding titles
- Using Pandas `.plot()` for quick charts




```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simple dataset for visualization
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 250, 300, 280, 350]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df["Month"], df["Sales"])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.show()
```

```python
plt.bar(df["Month"], df["Sales"])
plt.title("Monthly Sales - Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Example with random numbers
import numpy as np

data = np.random.randn(1000)  # 1000 random numbers
plt.hist(data, bins=30)
plt.title("Histogram of Random Data")
plt.show()
```

```python
# Plot directly from DataFrame
df.plot(x="Month", y="Sales", kind="line", title="Monthly Sales with Pandas")
plt.show()
```
